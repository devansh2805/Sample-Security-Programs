```python
from captcha.image import ImageCaptcha
from IPython.display import Image
import random
```

## Alphanumeric Captcha

```python
r1 = random.randint(1, 9)
captcha_list = []
for i in range(r1):
    r2 = random.randint(0,9)
    if r2 > 6:
        r3 = random.randint(0,9)
        captcha_list.append(str(r3))
    else:
        r3 = random.randint(0,25)
        captcha_list.append(chr(r3+97))
captcha_string = "".join(captcha_list)
```

```python
fonts = ['/usr/share/fonts/truetype/ubuntu/Ubuntu-B.ttf', '/usr/share/fonts/truetype/ubuntu/Ubuntu-BI.ttf']
captcha_image = ImageCaptcha(fonts=fonts)
captcha_image.write(captcha_string, 'out.png')
Image(filename='out.png')
```

```python
input_captcha_string = input('Enter Captcha: ')
if input_captcha_string == captcha_string:
    print('Correct Captcha')
else:
    print('Incorrect Captcha')
```

## Mathematical Captcha

```python
r1 = random.randint(25,360)
r2 = random.randint(20,100)
operator = random.choice(['*','/'])
answer = 0
if operator == '/':
    while(r1 % r2 != 0):
        r1 = random.randint(25,360)
        r2 = random.randint(20,100)
    answer = r1 / r2
else:
    answer = r1 * r2
math_captcha = str(r1) + operator + str(r2)
```

```python
captcha_image.write(math_captcha, 'mathcaptcha.png', )
Image(filename='mathcaptcha.png')
```

```python
input_math_answer = int(input("Enter Answer: "))
if input_math_answer == answer:
    print('Correct')
else:
    print('Incorrect')
```
